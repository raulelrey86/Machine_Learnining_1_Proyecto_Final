En conjunto, el enfoque PCA + clustering proporciona una herramienta exploratoria poderosa para analizar grandes volúmenes de texto, detectar temas latentes y comprender la estructura interna del conjunto de documentos sin requerir etiquetas predefinidas. </br></br>

2.2. **Preprocesamiento aplicado (limpieza, normalización, transformación, etc.):** </br></br>
Se realizó un proceso de preprocesamiento de los textos con el objetivo de limpiar, normalizar y transformar los datos del dataset Arxiv NLP en una representación adecuada para el análisis. Las etapas principales fueron las siguientes:</br>
**a)	Selección y combinación de campos relevantes:** </br>
Se utilizaron las columnas title y abstract de cada artículo, las cuales fueron concatenadas para formar un único texto representativo por documento, asegurando una mejor cobertura semántica del contenido.</br>
**b)	Limpieza de texto:** </br>
Se eliminaron caracteres especiales, signos de puntuación, números y espacios innecesarios. Además, se transformaron todos los textos a minúsculas para evitar duplicidad entre palabras con diferente capitalización (por ejemplo, “NLP” y “nlp”).</br>
**c)  Eliminación de palabras vacías (stopwords):** </br>
Se removieron las palabras de muy alta frecuencia y bajo valor informativo (como “the”, “and”, “of”, etc.) utilizando la lista de stopwords del idioma inglés, disponible en bibliotecas de procesamiento de texto como scikit-learn o NLTK.</br>
**d)	Tokenización y normalización:** </br>
Cada texto fue dividido en tokens (palabras individuales) y normalizado, eliminando variaciones irrelevantes de forma. Opcionalmente, se puede aplicar lemmatización o stemming para reducir las palabras a su raíz gramatical (por ejemplo, “running” → “run”).</br>
**e)	Vectorización con TF-IDF (Term Frequency – Inverse Document Frequency):** </br>
Los textos fueron transformados en vectores numéricos mediante el modelo TF-IDF, que pondera cada palabra según su frecuencia en el documento y su rareza en el corpus completo. Esta representación permite capturar la importancia relativa de los términos dentro de los artículos.</br>
**f)	Normalización de características:** </br>
Los vectores TF-IDF fueron escalados y convertidos en una matriz dispersa normalizada, lo que facilita su procesamiento posterior en algoritmos como PCA y K-Means, reduciendo el impacto de la magnitud de los valores.</br></br>
2.3. **Decisiones tomadas y razones:** </br></br>
Durante el desarrollo del análisis se tomaron diversas decisiones metodológicas con el propósito de optimizar la calidad de los resultados y la interpretabilidad de los grupos generados. </br>
A continuación, se describen las principales decisiones adoptadas y su respectiva justificación:</br>
**a)	Selección del enfoque no supervisado:** </br>
Se optó por un enfoque no supervisado debido a que el dataset Arxiv NLP no contiene etiquetas o categorías predefinidas que indiquen los temas exactos de cada artículo. Por tanto, técnicas como PCA y K-Means resultan apropiadas para descubrir estructuras y patrones ocultos dentro de los datos textuales sin requerir conocimiento previo.</br>
**b)	Uso del modelo TF-IDF para la representación de texto:** </br>
Se decidió emplear la representación TF-IDF (Term Frequency – Inverse Document Frequency) por su capacidad de resaltar las palabras más importantes de cada documento en relación con el conjunto total. A diferencia de los conteos simples, TF-IDF reduce la influencia de términos muy frecuentes, pero poco informativos (como “the” o “data”).</br>
**c)	Reducción de dimensionalidad con PCA:** </br>
Dado que la matriz TF-IDF posee miles de dimensiones (una por palabra), se aplicó Análisis de Componentes Principales (PCA) para reducir la complejidad del espacio de características, conservando la mayor parte de la varianza (información). Esto permitió visualizar los datos en 2D y mejorar la eficiencia de los algoritmos de clustering.</br>
**d)	Selección del algoritmo de clustering (K-Means):** </br>
Se eligió K-Means por ser un método robusto, eficiente y ampliamente utilizado para segmentar datos de alta dimensión. Además, su naturaleza iterativa permite ajustar el número de clusters (k) de forma controlada y comparar resultados visualmente. </br>
**e)	Elección del número de clusters (k = 10):** </br>
Se determinó un valor inicial de k = 10 como punto de partida para observar la distribución temática del conjunto de artículos. Este número puede ajustarse posteriormente con base en métricas de evaluación (como el Silhouette Score) o interpretabilidad de los resultados.

3.2. **Algoritmos de clustering aplicados (k-means, DBSCAN, etc.) y parámetros utilizados.** </br></br>

```python
# 1️⃣ Importar librerías

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt

# 2️⃣ Cargar dataset desde archivo CSV
df = pd.read_csv("data.csv")

# 3️⃣ Combinar columnas de texto (título + resumen)
df["text"] = df["Titles"].astype(str) + " " + df["Abstracts"].astype(str)

# 4️⃣ Vectorizar texto usando TF-IDF
#    Limitamos a 3000 características para eficiencia
vectorizer = TfidfVectorizer(max_features=3000, stop_words="english")
X = vectorizer.fit_transform(df["text"])

# 5️⃣ Aplicar TruncatedSVD (PCA optimizado para texto)
svd = TruncatedSVD(n_components=2, random_state=42)
X_pca = svd.fit_transform(X)

# 6️⃣ Crear DataFrame con los resultados del PCA
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_pca["Categories"] = df["Categories"]

# 7️⃣ Graficar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(df_pca["PC1"], df_pca["PC2"], s=10, alpha=0.6)
plt.title("Distribución de Datos en PCA (Titles + Abstracts)")
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# 8️⃣ (Opcional) Guardar resultados a CSV
df_pca.to_csv("pca_resultados.csv", index=False, encoding="utf-8")

print("✅ PCA completado. Resultados guardados en 'pca_resultados.csv'")
print(df_pca.head())
```

La proyección PCA revela que:

- El corpus de titles y abstracts de arXiv tiene una alta homogeneidad semántica general, con algunos documentos que se alejan del centro por tratar temas más específicos o inusuales.

- No se observan clústeres claramente definidos en el espacio PCA, por lo que para encontrar agrupaciones más significativas sería conveniente aplicar algoritmos de clustering (como k-means, DBSCAN o spectral clustering) sobre los componentes principales.

```python
# librerias
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Cargar dataset
df = pd.read_csv("data.csv")

# Combinar Titles + Abstracts
df["text"] = df["Titles"].fillna('') + " " + df["Abstracts"].fillna('')

# Vectorizar texto (TF-IDF)
vectorizer = TfidfVectorizer(stop_words='english', max_features=2000)
X = vectorizer.fit_transform(df["text"])

# Método del Codo (para elegir k)
inertia = []
K = range(2, 20)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(7,5))
plt.plot(K, inertia, 'bo-')
plt.title('Método del Codo para elegir k')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.show()

# Elegir k (por ejemplo el que ves en el "codo")
k = 10  # cámbialo según el gráfico anterior

# Aplicar K-Means definitivo
kmeans = KMeans(n_clusters=k, random_state=42)
df["cluster"] = kmeans.fit_predict(X)

# Reducir a 2D con PCA y graficar
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X.toarray())

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=df["cluster"], cmap='tab10', s=15)
plt.title(f"K-Means con k={k} (Titles + Abstracts)")
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.show()

# Mostrar resumen
print("\nCantidad de textos por cluster:")
print(df["cluster"].value_counts())
```

**Silhouette Score**</br></br>

Mide qué tan similares son los objetos dentro de un cluster en comparación con otros clusters. Su valor varía entre -1 y 1:</br></br>

- Valores cercanos a 1 indican clusters bien separados y cohesionados.</br>
- Valores cercanos a 0 indican solapamiento entre clusters.</br>
- Valores negativos sugieren que algunos puntos pueden estar mal asignados.</br></br>

Fórmula básica:</br>

$$ s(i) = \frac{b(i)-a(i)}{max(a(i),b(i))} $$

donde:

- **𝑎(𝑖) =** distancia promedio del punto **𝑖** con otros puntos de su cluster.
- **𝑏(𝑖) =** distancia promedio del punto **𝑖** con puntos del cluster más cercano diferente.
